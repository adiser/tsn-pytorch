```python
test = ['Tokyo', 'London', 'Rome', 'Donlon', 'Kyoto', 'Paris', 'Spari', 'Ahay', 'Haya']
```

```python
def find_pairs(test):
    var = [element.lower() for element in test]
    
    word_element_mappings = []
    
    for i in range(len(var)):
        word_element_mappings.append([var[i]])
        for j in range(len(var[i])):
            elements = []
            steps = len(var[i])
            
            start = j
            k = 0
            while k < steps:
                abs_index = j + k
                if abs_index >= steps:
                    abs_index = abs_index - steps
                elements.append(var[i][abs_index])
                
                k += 1
                
            word_element_mappings[i].append(elements)
            
    mappings = []
    taken = []
    for i in range(len(word_element_mappings)):
        if i not in taken:
            mappings.append([i])

            for j in range(len(word_element_mappings)):
                if i == j:
                    pass
                else:
                    for k in range(1, len(word_element_mappings[i])):
                        if word_element_mappings[i][k] in word_element_mappings[j]:
                            if sorted([i,j]) not in mappings:
                                mappings[i].append(j)
                                taken.append(j)
        else:
            mappings.append([])
        
    toreturn = []
    for i in range(len(mappings)):
        element = []
        for j in range(len(mappings[i])):
            element.append(var[mappings[i][j]].title())
        if element:
            toreturn.append(element)
        
    return toreturn
```

```python
find_pairs(test)
```
